Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')
df.head()
```

```python
X = df['Experience Years']
y = df['Salary']
```

```python
X = X.values.reshape(-1,1)
```

```python
y = y.values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2, random_state=365)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
coefficient = model.coef_[0][0]
coefficient
```

```python
intercept = model.intercept_[0]
intercept
```

```python
y_pred = model.predict(X)
```

```python
df['Predicted Salary'] = y_pred
```

```python
df
```

```python
plt.scatter(df['Experience Years'], df['Salary'], color='Blue', label='Original Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Original Salary VS Exp')
plt.show()
```

```python
plt.scatter(df['Experience Years'], df['Predicted Salary'], color='red', label='Predicted Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Predicted Salary VS Exp')
plt.show()
```

```python
plt.scatter(df['Experience Years'], df['Salary'], color='Blue', label='Original Salary')
plt.scatter(df['Experience Years'], df['Predicted Salary'], color='green', label='Predicted Salary')
plt.plot(df['Experience Years'], model.predict(X), color='red', label='Regression line')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
