```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
housing = fetch_california_housing()
```

```python
housing_data = pd.DataFrame(housing.data, columns=housing.feature_names)
```

```python
x = housing_data.drop(['MedInc'], axis=1)
y = housing_data['MedInc']
```

```python
x.shape
```

```python
y = y.values.reshape(-1,1)
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(x)
```

```python
y_pred
```

```python
housing_data['Predicted MedInc'] = y_pred
```

```python
plt.figure(figsize=(10,6))
plt.scatter(housing_data.index, y, label='Actual MedInc', color='blue', alpha=0.6)
plt.scatter(housing_data.index, y_pred, label='Predicted MedInc', color='red', alpha=0.6)
plt.xlabel('Index')
plt.ylabel('Median Income')
plt.title('Actual vs Predicted Median Income')
plt.legend()
plt.show()
```
