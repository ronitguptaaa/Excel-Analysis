Multiple Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Boston.csv')
df.head()
```

```python
df.columns
```

```python
X = df.drop(['MEDV'], axis=1)
```

```python
X.shape
```

```python
y = df['MEDV']
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.2,random_state=365)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
intercept = model.intercept_
intercept
```

```python
coefficient = model.coef_
coefficient
```

```python
y_pred = model.predict(X)
```

```python
df['Predictions'] = y_pred
```

```python
df.head()
```

```python
plt.scatter(df['AGE'], df['MEDV'], color='blue')
plt.show()
```

```python
plt.scatter(df['AGE'], df['Predictions'], color='green')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['AGE'], df['MEDV'], color='blue')
plt.scatter(df['AGE'], df['Predictions'], color='green')
plt.xlabel('Age')
plt.ylabel('Price')
plt.title('Age vs Price')
plt.legend()
plt.show()
```
